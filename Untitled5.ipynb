```python
from selenium import webdriver
from selenium.webdriver.common.keys import *
import time 
import pandas 
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import *
from selenium.webdriver.support.select import *
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
url="https://www.flipkart.com/"
driver=webdriver.Chrome()
driver.get(url)
content_click=driver.find_element(By.XPATH,'.//*[contains(concat( " ", @class, " " ), concat( " ", "Pke_EE", " " ))]')
content_click.send_keys("toys for kids")
content_click=driver.find_element(By.XPATH,'.//*[contains(concat( " ", @class, " " ), concat( " ", "_2iLD__", " " ))]')
content_click.click()

product_details = []
items=WebDriverWait(driver,10).until(
    EC.presence_of_all_elements_located((By.XPATH,'.//*[contains(concat( " ", @class, " " ), concat( " ", "slAVV4", " " ))]')))

for item in items:
    try:
        product = {
                'Name': item.find_element(By.XPATH, './/*[contains(concat( " ", @class, " " ), concat( " ", "wjcEIp", " " ))]').text,
                'Rating': item.find_element(By.XPATH, './/*[contains(concat( " ", @class, " " ), concat( " ", "XQDdHH", " " ))]').text,
                'Rating_count': item.find_element(By.XPATH, './/*[contains(concat( " ", @class, " " ), concat( " ", "Wphh3N", " " ))]').text,
                'New_Price': item.find_element(By.XPATH, './/*[contains(concat( " ", @class, " " ), concat( " ", "Nx9bqj", " " ))]').text,
                'Old_Price': item.find_element(By.XPATH, './/*[contains(concat( " ", @class, " " ), concat( " ", "yRaY8j", " " ))]').text,
                'Offer': item.find_element(By.XPATH, './/*[contains(concat( " ", @class, " " ), concat( " ", "UkUFwK", " " ))]//span').text,
            
        }
        product_details.append(product) 
    except Exception as e:
        print(f"Error extracting product details: {e}")
print(product_details)
```

```python
import pandas as pd
df=pd.DataFrame(product_details
               )
df=df.drop_duplicates()
```
